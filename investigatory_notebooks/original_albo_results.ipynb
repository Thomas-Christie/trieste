```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

### LSQ

```python
original_albo_lsq_results = []
for run in range(1, 101):
    with open(f"../results/original_albo/lsq/slack_al_optim/data/run_{run}_results.json", "rb") as fp:
        all_run_data = json.load(fp)
        run_results = all_run_data["prog"][:45]
        i = 0
        while run_results[i] == "Inf":
            i += 1
        for j in range(i):
            run_results[j] = run_results[i]
        original_albo_lsq_results.append(run_results)
original_albo_lsq_results = np.array(original_albo_lsq_results)
original_albo_lsq_results_mean = np.mean(original_albo_lsq_results, axis=0)
original_albo_lsq_results_median = np.median(original_albo_lsq_results, axis=0)
```

```python
updated_albo_lsq_results = []
for run in range(1, 101):
    with open(f"../results/original_albo/lsq/updated_rho_slack_al_optim/data/run_{run}_results.json", "rb") as fp:
        all_run_data = json.load(fp)
        run_results = all_run_data["prog"][:45]
        i = 0
        while run_results[i] == "Inf":
            i += 1
        for j in range(i):
            run_results[j] = run_results[i]
        updated_albo_lsq_results.append(run_results)
updated_albo_lsq_results = np.array(updated_albo_lsq_results)
updated_albo_lsq_results_mean = np.mean(updated_albo_lsq_results, axis=0)
updated_albo_lsq_results_median = np.median(updated_albo_lsq_results, axis=0)
```

```python
plt.plot(original_albo_lsq_results_median, label="Original Penalty Initialisation (Median)")
plt.plot(updated_albo_lsq_results_median, label="Updated Penalty Initialisation (Median)")
plt.legend()
plt.savefig("lsq_median.png", dpi=200)
plt.show()
```

```python
plt.plot(original_albo_lsq_results_mean, label="Original Penalty Initialisation (Mean)")
plt.plot(updated_albo_lsq_results_mean, label="Updated Penalty Initialisation (Mean)")
plt.legend()
plt.savefig("lsq_mean.jpg", dpi=200)
plt.show()
```

```python
original_penalties = []
for run in range(1, 101):
    with open(f"../results/original_albo/lsq/slack_al_optim/data/run_{run}_results.json", "rb") as fp:
        all_run_data = json.load(fp)
        initial_penalty = all_run_data["rho"][0]
        original_penalties.append(initial_penalty)
```

```python
updated_penalties = []
for run in range(1, 101):
    with open(f"../results/original_albo/lsq/updated_rho_slack_al_optim/data/run_{run}_results.json", "rb") as fp:
        all_run_data = json.load(fp)
        initial_penalty = all_run_data["rho"][0]
        updated_penalties.append(initial_penalty)
```

```python
print(f"Original Penalties Mean: {np.mean(original_penalties)} STD: {np.std(original_penalties)}")
print(f"Updated Penalties Mean: {np.mean(updated_penalties)} STD: {np.std(updated_penalties)}")
```
